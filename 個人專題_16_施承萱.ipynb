```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
data = pd.read_csv("/content/drive/MyDrive/AI人才應用/Dataset_2023-24 NBA Player Stats.csv")
data.head(10)
```

```python
data.info()
```

```python
Data_isnull = data.isnull().sum()
df = pd.DataFrame(data = Data_isnull)
new_columns_name = {0 : '有缺失值欄位的總和'}
df.rename(new_columns_name, axis = 1)
```

```python
sns.relplot(data=data, x="Age", y="PTS")
```

```python
sns.relplot(data=data, x="Pos", y="PTS")
```

```python
sns.catplot(data=data, x="Pos", y="PTS", kind="swarm", s=8)
sns.displot(data=data, x="PTS", hue="Pos", kde=True)
sns.catplot(data=data, x="Pos", y="PTS", kind="box")
sns.catplot(data=data, x="Pos", y="PTS", kind="boxen")
sns.catplot(data=data, x="Pos", y="PTS", kind="violin")
sns.catplot(data=data, x="Pos", y="PTS", kind="point")
```

```python
mean_points = data.groupby("Pos")["PTS"].mean().reset_index(name="MeanPoints")
ax = sns.barplot(mean_points, x="MeanPoints", y="Pos", color="skyblue")
```

```python
# Who are the top 5 players with the most points per game (PTS)?
top_5_scorers = data.nlargest(5, 'PTS')['Player']
```

```python
# Bar plot of top 5 scorers
plt.figure(figsize=(10,6))
sns.barplot(x=top_5_scorers, y=data.loc[top_5_scorers.index, 'PTS'])
plt.title('Top 5 Players with Most Points per Game')
plt.xticks(rotation=90)
plt.xlabel('Player')
plt.ylabel('Points per Game')
plt.show()
```
